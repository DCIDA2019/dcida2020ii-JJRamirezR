```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import binom
from scipy.stats import poisson
```

# Ejercicio 1
Homework: Find the cumulative distribution function for the Gaussian Distribution, and reproduce the plots. Choose a random number between 0 and 1, and use the CDF to assign the corresponding value of H0. Generate as many as you want, and make the histogram of H0 to verify you did it right. Use a mean of 70 and a sigma=2

```python
#Datos
mean=70
sigma=2
```

```python
x=np.random.random(10000)
H0=norm.ppf(x,mean,sigma)
P0=1/(sigma*np.sqrt(2*np.pi))*np.exp(-((H0-mean)*(H0-mean))/(2*sigma*sigma))
```

```python
plt.hist(H0)
```

```python
plt.plot(H0,x,'g.')
plt.axhline(y=0.5, xmin=0, xmax=.5)
plt.axvline(x=70, ymin=0, ymax=.5)
plt.ylabel('$ F(H_0) $')
plt.xlabel('$ H_0 $')
plt.title('Distribición Normal')
plt.show()
```

```python
plt.plot(H0,P0,'r.')
plt.ylabel('$ P(H_0) $')
plt.xlabel('$ H_0 $')
plt.title('Distribición Normal')
plt.show()
```

## Distribución de Poisson 

```python
#Parámetros de forma
mu1=2.5
mu2=3.6
mu3=4.3
# Distribución
poi1 = poisson(mu1) 
poi2 = poisson(mu2)
poi3 = poisson(mu3)

xp1 = np.arange(poi1.ppf(0.01),poi1.ppf(0.99))
xp2 = np.arange(poi2.ppf(0.01),poi2.ppf(0.99))
xp3 = np.arange(poi3.ppf(0.01),poi3.ppf(0.99))

ppmf1=poi1.pmf(xp1)
ppmf2=poi2.pmf(xp2)
ppmf3=poi3.pmf(xp3)

pcdf1 = poi1.cdf(xp1) 
pcdf2 = poi2.cdf(xp2)
pcdf3 = poi3.cdf(xp3)
```

```python
plt.plot(xp1, ppmf1, '--', label='$\mu$=2.5')
plt.plot(xp2,ppmf2, '--', color='g', label='$\mu$=3.6')
plt.plot(xp3,ppmf3, '--', color='r', label='$\mu$=4.3')
plt.legend()
plt.title('Distribución de Poisson PDF')
plt.ylabel('$P(x)$')
plt.grid()
plt.xlabel('x')
plt.show()
```

```python
plt.plot(xp1, pcdf1, '--', label='$\mu$=2.5')
plt.plot(xp2,pcdf2, '--', color='g', label='$\mu$=3.6')
plt.plot(xp3,pcdf3, '--', color='r', label='$\mu$=4.3')
plt.legend()
plt.title('Distribución de Poisson CDF')
plt.ylabel('$F(x)$')
plt.grid()
plt.xlabel('x')
plt.show()
```

## Distribución Binomial

```python
n1, p1 = 30, 0.4 # parametros de forma
n2, p2 = 35, 0.4 # parametros de forma
n3, p3 = 40, 0.4 # parametros de forma
# Distribución
bin1 = binom(n1, p1) 
bin2 = binom(n2, p2)
bin3 = binom(n3, p3)

x1 = np.arange(bin1.ppf(0.01),bin1.ppf(0.99))
x2 = np.arange(bin2.ppf(0.01),bin2.ppf(0.99))
x3 = np.arange(bin3.ppf(0.01),bin3.ppf(0.99))

pmf1=bin1.pmf(x1)
pmf2=bin2.pmf(x2)
pmf3=bin3.pmf(x3)

cdf1 = bin1.cdf(x1) 
cdf2 = bin2.cdf(x2)
cdf3 = bin3.cdf(x3)
```

```python
plt.plot(x1, pmf1, '--', label='n=30,p=0.4')
plt.plot(x2,pmf2, '--', color='g', label='n=35, p=0.4')
plt.plot(x3,pmf3, '--', color='r', label='n=40, p=0.4')
plt.legend()
plt.title('Distribución Binomial PDF')
plt.ylabel('$P(x)$')
plt.grid()
plt.xlabel('x')
plt.show()
```

```python
plt.plot(x1, cdf1, '--', label='n=30,p=0.4')
plt.plot(x2,cdf2, '--', color='g', label='n=35, p=0.4')
plt.plot(x3,cdf3, '--', color='r', label='n=40, p=0.4')
plt.legend()
plt.title('Distribución Binomial CDF')
plt.ylabel('$F(x)$')
plt.grid()
plt.xlabel('x')
plt.show()
```
