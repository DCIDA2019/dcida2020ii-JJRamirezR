```python
#Librerías
# import numpy as np
import os
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
# Los diferentes url de la tablas escogidas de la página indexmundi

#North America
urlNA1='https://www.indexmundi.com/map/?t=0&v=69&r=na&l=en'
urlNA2='https://www.indexmundi.com/map/?t=0&v=26&r=na&l=en'
urlNA3='https://www.indexmundi.com/map/?t=0&v=2227&r=na&l=en'
urlNA4='https://www.indexmundi.com/map/?t=0&v=2226&r=na&l=en'
urlNA5='https://www.indexmundi.com/map/?t=0&v=30&r=na&l=en'
#South America
urlSA1='https://www.indexmundi.com/map/?t=0&v=69&r=sa&l=en'
urlSA2='https://www.indexmundi.com/map/?t=0&v=26&r=sa&l=en'
urlSA3='https://www.indexmundi.com/map/?t=0&v=2227&r=sa&l=en'
urlSA4='https://www.indexmundi.com/map/?t=0&v=2226&r=sa&l=en'
urlSA5='https://www.indexmundi.com/map/?t=0&v=30&r=sa&l=en'
#Central America
urlCA1='https://www.indexmundi.com/map/?t=0&v=69&r=ca&l=en'
urlCA2='https://www.indexmundi.com/map/?t=0&v=26&r=ca&l=en'
urlCA3='https://www.indexmundi.com/map/?t=0&v=2227&r=ca&l=en'
urlCA4='https://www.indexmundi.com/map/?t=0&v=2226&r=ca&l=en'
urlCA5='https://www.indexmundi.com/map/?t=0&v=30&r=ca&l=en'
```

```python
#Función para crear tabla:
def tabla(file,url1,url2,url3,url4,url5):
    if not os.path.isfile(file):
        html1 = requests.get(url1).content
        html2 = requests.get(url2).content
        html3 = requests.get(url3).content
        html4 = requests.get(url4).content
        html5 = requests.get(url5).content
        df_list1 =pd.read_html(html1, header=0)
        df1 = df_list1[1]
        df1=df1.drop(['Year'],axis=1)
        df_list2 =pd.read_html(html2, header=0)
        df2 = df_list2[1]
        df2=df2.drop(['Year'],axis=1)
        df_list3 =pd.read_html(html3, header=0)
        df3 = df_list3[1]
        df3=df3.drop(['Year'],axis=1)
        df_list4 =pd.read_html(html4, header=0)
        df4 = df_list4[1]
        df4=df4.drop(['Year'],axis=1)
        df_list5 =pd.read_html(html5, header=0)
        df5 = df_list5[1]
        df5=df5.drop(['Year'],axis=1)
        df=pd.merge(df1,df2)
        df=pd.merge(df,df3)
        df=pd.merge(df,df4)
        df=pd.merge(df,df5)
        df.to_csv(file)
    else:
        df = pd.read_csv(file)
        df=df.drop(['Unnamed: 0'],axis=1)
    return df      
```

```python
#Archivo con la información de North America
fileNA='NAmerica.csv'
NA=tabla(fileNA,urlNA1,urlNA2,urlNA3,urlNA4,urlNA5)
#Archivo con información de Central America
fileCA='CAmerica.csv'
CA=tabla(fileCA,urlCA1,urlCA2,urlCA3,urlCA4,urlCA5)
#Archivo con información de South America
fileSA='SAmerica.csv'
SA=tabla(fileSA,urlSA1,urlSA2,urlSA3,urlSA4,urlSA5)
```

```python
#Tabla con todos los valores de America
A=pd.concat([NA,CA,SA])
A.index=range(A.shape[0])
A
```
