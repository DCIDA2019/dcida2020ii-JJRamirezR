# Descripción del programa
## En este programa se usan los comandos concat y merge 

```python
#importando las librerías
import pandas as pd
import requests
```

```python
#función para generar tabla de página web
def tabla(url):
 html = requests.get(url).content
 df_list =pd.read_html(html, header=0)
 df=df_list[1] 
 return df 
```

```python
#Tabla birth rate North America
url='https://www.indexmundi.com/map/?t=0&v=25&r=na&l=en'
birth=tabla(url)
birth
```

```python
#Tabla population North America
url2='https://www.indexmundi.com/map/?t=0&v=21&r=na&l=en'
pop=tabla(url2)
pop
```

```python
#Tabla birth rate South America
url3='https://www.indexmundi.com/map/?t=0&v=25&r=sa&l=en'
birthS=tabla(url3)
birthS
```

```python
#Tabla Net migration rate North America
#url4='https://www.indexmundi.com/map/?t=0&v=27&r=na&l=en'
#mig=tabla(url4)
#mig
```

# Uso de concat

```python
# tabla birth rate North America + South America
pd.concat([birth,birthS])
```

```python
# tabla birth rate + population North America
pd.concat([birth,pop],axis='columns')
```

# Uso de Merge y Join

```python
#tabla birth rate + population North America
pd.merge(birth,pop)
```

```python
#pd.merge(birth,birthS, how='outer')
pd.merge(pd.merge(birth,pop),birthS, how='outer')
```
